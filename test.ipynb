```python
import pandas as pd
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
df_train = pd.read_csv('train.csv')
df_train
```

```python
df_test = pd.read_csv('test.csv')
df_test
```

```python
df_concat = pd.concat([df_train,df_test])
df_concat
```

```python
def data_cleaning(df:pd.DataFrame, column:str):
    list_val = df[column].value_counts()[df[column].value_counts(normalize=True)<0.01].index
    for value in list_val:
        df[column] = df[column].replace(value,'other')
    return df
```

```python
list_col=['cap-shape', 'cap-surface', 'cap-color',
       'does-bruise-or-bleed', 'gill-attachment', 'gill-spacing', 'gill-color','stem-root', 'stem-surface', 'stem-color',
       'veil-type', 'veil-color', 'has-ring', 'ring-type', 'spore-print-color',
       'habitat', 'season']
tmp = df_concat
for column in list_col:
    tmp = data_cleaning(tmp,column)
tmp
```

```python
tmp.info()
```

```python
x = pd.get_dummies(tmp[:len(df_train)].drop(columns=['id','class']),columns=list_col)
x
```

```python
y = pd.get_dummies(tmp[:len(df_train)]['class'])
y = y['p']
y
```

```python
clf = HistGradientBoostingClassifier()
clf.fit(x,y)
```

```python
x = pd.get_dummies(tmp[len(df_train):].drop(columns=['id','class']),columns=list_col)
x
```

```python
pred = clf.predict(x)
pred
```

```python
df_result = df_test['id']
df_result=pd.concat([df_result,pd.DataFrame(pred,columns=['class'])],axis=1)
df_result = df_result.replace(False,'e')
df_result = df_result.replace(True,'p')
df_result
```

```python
df_result.to_csv('result.csv',index=False)
```
